```python
import matplotlib.pyplot as plt

# Data
models = [
    "SRCNN", "FSRCNN", "EDRN", "SRDRN", "U-Net",
    "AU-Net", "U-GAN", "AU-GAN", "DL-ENS", "GAN-ENS"
]
drop_percent = [
    -5.52, -4.46, -5.37, -4.76, -4.91,
    -4.80, -3.62, -4.07, -4.91, -4.41
]

# Reverse the data for top-down order
models = models[::-1]
drop_percent = drop_percent[::-1]

# Plot
plt.figure(figsize=(8, 6))
# plt.grid(axis='y', linestyle='--', alpha=0.6, zorder=0)
bars = plt.barh(models, drop_percent, color='teal', edgecolor='black', zorder=1)

plt.xlabel("Drop in KGE Score (%)", fontsize=13, fontfamily='serif',)
plt.title("PERFORMANCE DROP (%)", fontsize=14, fontfamily='serif', fontweight='semibold')
plt.yticks(fontsize=13)
plt.xticks(fontsize=13)


# Add value + model name inside each bar
for bar, model_name in zip(bars, models):
    xval = bar.get_width()
    yval = bar.get_y() + bar.get_height() / 2
    text = f'{xval:.2f}% '
    plt.text(xval + 0.8, yval, text, va='center', ha='right', color='yellow', fontweight="semibold", fontsize=14)

# Adjust layout and save
plt.tight_layout()
SAVE_DIR = '/home/midhunm/AI4KLIM/EXPMNTS/P02A_DeepDown_Generalization/ANALYSE/RESULTS'
# plt.savefig(f'{SAVE_DIR}/Figure02_TransferSkillDrop.jpg', format='jpg', dpi=500, bbox_inches='tight', facecolor='w')
plt.show()
```
