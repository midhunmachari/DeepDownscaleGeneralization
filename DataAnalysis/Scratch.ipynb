```python
import tensorflow as tf

# --- Include your attention_gate function here (as you've defined above) ---

def get_layer_name(name):
    return name

def attention_gate(x, g, inter_channels, name='attention'):
    theta_x = tf.keras.layers.Conv2D(inter_channels, kernel_size=1, strides=1, padding='same', name=get_layer_name(f'{name}_attn_theta_x'))(x)
    phi_g = tf.keras.layers.Conv2D(inter_channels, kernel_size=1, strides=1, padding='same', name=get_layer_name(f'{name}_attn_phi_g'))(g)
    f = tf.keras.layers.Activation('relu', name=get_layer_name(f'{name}_attn_relu'))(theta_x + phi_g)
    psi = tf.keras.layers.Conv2D(1, kernel_size=1, strides=1, padding='same', name=get_layer_name(f'{name}_psi'))(f)
    psi = tf.keras.layers.Activation('sigmoid', name=get_layer_name(f'{name}_attn_sigmoid'))(psi)
    out = tf.keras.layers.Multiply(name=get_layer_name(f'{name}_attn_multiply'))([x, psi])
    return out

# Define dummy input shapes
input_x = tf.keras.Input(shape=(32, 32, 64), name="encoder_input")  # skip connection
input_g = tf.keras.Input(shape=(32, 32, 128), name="decoder_input")  # decoder signal

# Apply attention gate
attended_x = attention_gate(input_x, input_g, inter_channels=32, name='demo')

# Create the model
model = tf.keras.Model(inputs=[input_x, input_g], outputs=attended_x, name="DummyAttentionGateModel")
model.summary()

model.save("/home/midhunm/attention_fate.keras")
```
